```python
ranks = pd.get_dummies(df['rank'])
```

```python
ranks.head()
```

```python
X = pd.concat([df[['gre', 'gpa']], ranks], axis=1)
X1 = df[['gre', 'gpa', 'rank']]
# X = df[['gre', 'gpa']]

y = df['admit']
```

```python
from sklearn.feature_selection import chi2
```

```python
scores, pvalues = chi2(X, y)
print(['{:.6f}'.format(score) for score in scores])
print(['{:.6f}'.format(pvalue) for pvalue in pvalues])
```

```python
from sklearn.linear_model import LogisticRegression, RidgeClassifier, SGDClassifier

reg = LogisticRegression(C=1.)
reg1 = LogisticRegression(C=10000)
X_train, y_train, X_test, y_test = X[:-100], y[:-100], X[-100:], y[-100:]
clf = reg.fit(X_train, y_train)
clf1 = reg1.fit(X1, y)
```

```python
print(clf.intercept_)
print(clf.coef_)
```

```python
print(clf1.intercept_)
print(clf1.coef_)
```

```python
from sklearn.metrics import accuracy_score

y_ = clf.predict(X_test)
accuracy_score(y_, y_test)
```

```python
y1_ = clf1.predict(X1)
accuracy_score(y1_, y)
```

```python
ridge = RidgeClassifier()
clfr = ridge.fit(X_train, y_train)
yr_ = clfr.predict(X_test)
accuracy_score(yr_, y_test)
```

```python
sgd = SGDClassifier(loss='log')
clfs = sgd.fit(X_train, y_train)
ys_ = clfs.predict(X_test)
accuracy_score(ys_, y_test)
```

```python
from sklearn.feature_selection import f_regression
fs, ps = f_regression(X, y)
fs = ['{:.5f}'.format(f) for f in fs]
ps = ['{:.5f}'.format(p) for p in ps]
print('f-values: %s' % fs)
print('p-values: %s' % ps)
```
