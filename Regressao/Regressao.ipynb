```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import statsmodels.formula.api as smf
```

```python
base_path = 'Aula 10'
```

# Ex17

```python
df = pd.read_stata(os.path.join(base_path, 'CRIME2-Kayo.dta'), preserve_dtypes=False)
```

```python
df.head()
```

```python
df.columns
```

```python
df[['crmrte', 'unem', 'loffic']].describe()
```

```python
df87 = df[df['year'] == 87]
```

```python
model = smf.ols(formula ='crmrte ~ unem', data=df87)
results = model.fit()
print(results.summary())
```

# Ex18

```python
model = smf.ols(formula ='crmrte ~ d87 + unem', data=df)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula ='crmrte ~ unem + loffic', data=df)
results = model.fit()
print(results.summary())
```

```python
crmrte_list = df['crmrte'].tolist()[1:]
crmrte_list.append(df['crmrte'].tolist()[-1])
print(crmrte_list)
```

```python
print(len(df))
print(len(crmrte_list))
```

```python
df2 = df[['crmrte', 'unem', 'loffic']]
```

```python
se = pd.Series(crmrte_list)
df2['crmrtedef'] =  se.values
```

```python
df2.head()
```

```python
model = smf.ols(formula ='crmrtedef ~ unem + loffic', data=df2)
results = model.fit()
print(results.summary())
```
